# Density plot

Defination : A density plot (or Kernel Density Estimate (KDE) plot) is a data visualization technique that shows the probability density function (PDF) of a continuous variable

When to Use:
Best for visualizing the probability density of a continuous variable (i.e., showing how data points are distributed)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate random data
'''
np.random.normal(a,b,n) -> gives total n datapoints which follows normal distribution with mu=a and sd=b
'''
data = np.random.normal(0,4,100000)  # 1000 data points from a normal distribution

# Create the density plot
'''
kdeplot() takes some arguments:
color='red' - Sets the color of the histogram bars
fill=bool(True/False) - Fills color in the area under curve
cumulative=bool (True/False) - Gives the cumulative distribution plot
bw_method=scalar (0.01) - Method for determining the smoothing bandwidth to use 
'''
#Density Plot using Kernel Density Estimation with Seaborn
sns.kdeplot(data, fill=True,color='Red')

# Plot and title
plt.title("Density Plot of Normal Distribution(0,4)")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()

#Density Plot using Matplotlib

counts, bin_edges = np.histogram(data,bins=15,density=True)
midpoints = (bin_edges[:-1] + bin_edges[1:]) / 2
plt.plot(midpoints, counts, color='red', label='Midpoints')
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Density plot using Matplotlib")
plt.fill_between(midpoints,counts,alpha=0.6)

plt.show()
```

# Cumulative Density Plot

```python
#Generation of Data points
'''
np.random.uniform(a,b,n) -> gives total n datapoints which follows uniform distribution with parameters a and b
'''
uni_dist=np.random.uniform(1,3,10000)

#Cumulative Density Plot using KDE method with Seaborn
sns.kdeplot(uni_dist,cumulative=True,fill=True,color='Green')

#Plot and Title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title("Cumulative Density plot of Uniform distribution(1,3)")

plt.show()
```

# Using Real Life Data

```python
#Seaborn has a lot of Popular data sets in its database which we can simply import
iris_dataset=sns.load_dataset('iris')
print(iris_dataset.head(3))     #Data Visualization

#Plot distributions for each column of a wide-form dataset
sns.kdeplot(iris_dataset,fill=True)

#Plot and Title
plt.xlabel('Length/Width (cm)')
plt.ylabel("Density")
plt.title("KDE Plot of Sepal and Petal Dimensions for Iris Dataset")

plt.show()
```

* # Pros:
1. Density plots provide a continuous and smooth estimate of the distribution
2. Density plots are great for overlaying multiple distributions in one plot, making it easy to compare multiple datasets or variables visually
3. Since density plots estimate a probability density function (PDF), they are ideal for visualizing continuous variables

* # Cons:
1. For small datasets, density plots can be misleading. The smoothing might hide the true variability of the data, and peaks might appear where there are no actual data points.
2. Density plots are generally better suited for continuous data. Using them with discrete or categorical data can be confusing and misleading, as they assume a continuous probability distribution
3. When overlaying many distributions, the plot can become cluttered. Too many density curves can overlap
