# Box Plot

Defination :

A box plot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on five summary statistics: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It also shows potential outliers. The box represents the interquartile range (IQR), which contains the middle 50% of the data, and the "whiskers" extend to the minimum and maximum values within a specified range. Any points outside the whiskers are considered outliers

When to use:

Box plot is mainly used to identify and visualize the outlier in a perticular dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
'''
plt.boxplot() takes some arguments:
x: if x is 1D data then only 1 box plot will appear 
   but if x is 2D array then boxplot creates 2 seperate boxplots
notch: bool(True/False) Shows a notch, which helps to understand median 
whis: lets us choose the whisker size (def=1.5)

and using dictionaries boxprops(), medianprops(), whiskerprops() and capprops()
we can edit the box plot color and other aspects
'''
# Generating Data
x=np.random.normal(8,25,30)
print(x)

# Boxplot and Title
plt.boxplot(x, notch=False, whis=0.9, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            medianprops=dict(color='red', linewidth=2),
            capprops=dict(color='black'))

plt.title(r"Boxplot of Normally Distributed Data with $\mu=8, \sigma=25$")
plt.ylabel("Values")

plt.show()
```

# Using Real Life Data

```python
# Importing Data from csv file using Pandas
data=pd.read_csv('titanic.csv')
#print(data.head(3))

# Storing data of interest
age=data['Age'].dropna()
print(age.head())

# Calculate the median and quartiles
median_age = np.median(age)
q1 = np.percentile(age, 25)
q3 = np.percentile(age, 75)

#Box plot and title
sns.boxplot(age,color='Brown',width=0.3)
plt.ylabel("Age")
plt.title("Boxplot of Passenger Ages on the Titanic")
#plt.axhline(median_age)

plt.show()

print(f"The median is {median_age}, and the 1st and 3rd quartiles are {q1}, {q3} respectively")

```

From the above box plot we can infer:
* The median passenger age is 28 years
* The first quartile is approximately 20 indicating 25% of the passengers were younger than 20
* The third quartile is approximately 38 indicating 75% of the passengers were younger than 38

# Pros:
1. Box plots provide a clear summary of a dataset's distribution, including key statistics like the median, quartiles, and range
2. Box plots are excellent for identifying outliers
3. Box plots are ideal for comparing distributions across different categories or groups

# Cons: 
1. Box plots do not show individual data points (except for outliers)
2. When used with small datasets, box plots may be less informative and might not show meaningful insights
3. Box plots don’t provide information about the mode of the data
