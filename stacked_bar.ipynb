# STACKED BAR CHART

A stacked bar chart is a type of bar chart where each bar is divided into multiple segments, representing different categories. These segments are "stacked" on top of each other to show how a total is divided into parts. Itâ€™s useful when you want to compare both the overall size of categories and the composition of these categories across multiple groups or time periods.
    
* Data Source:	Kaggle

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data=pd.read_csv(r"D:\ISI\Acads\Assignments\Car_sales_data.csv")
#data.head()
categories=list(set(data['Body Style']))
#print(categories)
pt=data.pivot_table(index='Gender',columns='Body Style',aggfunc='size',fill_value=0)
print(pt)
pt=pt.values.tolist()
print(pt)
subcat1=pt[0]
subcat2=pt[1]
print(subcat1)
print(subcat2)
fig,ax=plt.subplots()
bar_width=0.5
ax.bar(categories,subcat1,label='Female',width=bar_width,color='Green')
ax.bar(categories,subcat2,bottom=subcat1,label='Male',width=bar_width,color='Purple')
#print(ax.get_xticklabels())
for label in ax.get_xticklabels():
    label.set_fontsize(9)
    
ax.set_xlabel("Body Style",fontdict={'fontweight':10,'fontweight':'bold'})
ax.set_ylabel("No. of customers",fontdict={'fontweight':10,'fontweight':'bold'})
ax.set_title("Stacked bar chart of Gender-wise distribution of Car Types",fontdict={'fontweight':15,'fontweight':'bold'})
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the data
categories = ['A', 'B', 'C', 'D']  # Categories for the X-axis (e.g., regions or years)
subcat_1 = [10, 20, 30, 40]  # Data for the first subcategory (e.g., sales in category 1)
subcat_2 = [15, 25, 10, 30]  # Data for the second subcategory (e.g., sales in category 2)
subcat_3 = [25, 15, 20, 10]  # Data for the third subcategory (e.g., sales in category 3)

# Step 2: Create a figure and axis for plotting
fig, ax = plt.subplots()

# Step 3: Create the stacked bar chart
bar_width = 0.5  # Width of each bar

# Plot the first subcategory
ax.bar(categories, subcat_1, label='Category 1', width=bar_width,color='Green')

# Plot the second subcategory on top of the first (stacking)
ax.bar(categories, subcat_2, bottom=subcat_1, label='Category 2', width=bar_width)

# Plot the third subcategory on top of the second (stacking)
ax.bar(categories, subcat_3, bottom=np.array(subcat_1) + np.array(subcat_2), label='Category 3', width=bar_width)

# Step 4: Add labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Chart Example')

# Step 5: Add a legend
ax.legend()

# Step 6: Display the chart
plt.show()
```
