# Pie Chart

Defination : 
A pie chart is a circular graph that is divided into slices to represent proportions or percentages of a whole. Each slice corresponds to a category, and its size is proportional to the quantity or percentage it represents. The entire circle represents 100% of the data

When to Use: 
Pie charts are best used to show the relative proportions or percentages of different categories in a dataset.Pie charts can be useful when you want to emphasize one or a few dominant categories in the data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
'''
matplotlib.pyplot.pie() takes some arguments:
colors- A sequence of colors through which the pie chart will cycle
autopct- The label will be placed inside the wedge. If autopct is a format string
labels- Labels the portions accordingly
radius- Set the radius of the circle
'''

#Example Data
labels = ['Apple','Lichi','Guava','Papaya','Mango']
sizes = [15,13,5,20,47]

# Pie chart and Label
plt.pie(sizes,autopct='%1.1f%%',radius=2)
plt.title("Distribution of Fruit Preferences", fontsize=16,loc='center')

# Optionally add a legend if needed
plt.legend(labels, loc='best', fontsize=12)

# Show plot
plt.show()
```

# Using Real Life Data

```python
# Importing Data from csv file using pandas
data=pd.read_csv('crime_dataset_india.csv')
#print(data.head(2))

# Storing Data of Interest
crime_city=data['City']

# To count individual Occerences 
occr_crime_city=crime_city.value_counts()
#print(occr_crime_city)

# Filtering cities with occurrences greater than 1100
filtered_crime_city = occr_crime_city[occr_crime_city > 1100]
print(filtered_crime_city)

# To generate random RGB 
colors = np.random.rand(12,3)   #This gives 12 random RGB combinations
#Ex= [0 1 0] mean full green

# Pie Chart and Label
plt.pie(filtered_crime_city,
        labels=filtered_crime_city.index,
        autopct='%1.1f%%',radius=1.5,
        colors=colors,
        textprops={'fontsize': 14})

plt.show()
```

From this we can infer that:
* Delhi has the greatest proportion of crimes(17.7%)
* Mumbai also has a good amount of crime rate (14.5%)
* Cities like Surat, Kanpur, Lucknow has relatively lower crime rate ranging from 3 to 5 %

# Pros: 
1. Pie charts provide a straightforward way to visualize parts of a whole, making them easy to interpret at a glance
2. When dealing with a small number of categories, pie charts are effective for showing proportions, percentages
3. Pie charts highlight the relative sizes of different components, which can be useful for comparing the significance of various parts of the dataset

# Cons:
1. Pie charts become cluttered and hard to interpret when there are too many categories
2. It can be challenging to accurately compare slices, especially if they are similar in size
3. Pie charts are not precise tools for showing exact values
