# Histogram

Definition:
A plot that groups data into bins and shows the frequency of data points within each bin.

When to Use:
Best for showing the distribution of a dataset (e.g., exam scores).

```python
import numpy as np
import matplotlib.pyplot as plt

'''hist() returns a tuple of 2 arrays: the first array is frequencies and the second array consists of the divisions'''

'''
hist() takes a some arguments:
color='red' - Sets the color of the histogram bars.
bars=30 - Determines the number of equal-width bins in the range.
alpha=0.5 - Controls the transparency of the histogram bars.
edgecolor='black' - Determines the color of the edges of each bar.
'''

#Generating Data Points for Histogram
norm_dist=np.random.normal(2,5,10000)       # Gives 10000 data points which follow normal distribution with mean=2 and sd=5

#Histogram of the data points
plt.hist(norm_dist,bins=30,edgecolor='black',alpha=0.6)

#Customize plot
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(r"Normal Distribution Histogram ($\mu = 2 , \sigma=5$)")
plt.xticks(np.arange(-20,21,4))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.random.exponential(0.7,10000)

# Create a step histogram
plt.hist(x, bins=30,edgecolor='black',color='orange')

# Customize plot
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(r'Exponential Distribution Histogram ($\lambda = 0.7$)')

plt.show()
```

# Using Real Life Data

```python
import pandas as pd

# Importing data from CSV file using pandas
data = pd.read_csv('StudentPerformanceFactors.csv')

# Storing data of interest
hour_std = data['Hours_Studied'].to_numpy()
print(hour_std)

# Plotting histogram of 'Hours_Studied'
plt.hist(hour_std, bins=30, edgecolor='black',alpha=0.7,color='red') 
plt.xlabel('Hours Studied')
plt.ylabel('Frequency')
plt.title('Histogram of Hours Studied')

plt.show()

#For Calculations of Statistical Measures
mean_hours=np.mean(hour_std)
med_hours=np.mean(hour_std)


print(f"Mean of Hours Studied: {mean_hours}")
print(f"Median of Hours Studied: {med_hours}")

```

* ## Pros:
1. Histograms clearly display the shape of the data distribution
2. Work well for summarizing large amounts of data
3. By visualizing the data in bins, histograms can give insights into the mean, median, mode, and the spread of the data

* ## Cons:
1. The appearance of a histogram can change dramatically based on the bin size chosen
2. Histograms group data into bins, which can lead to a loss of precise information
3. Histograms are only suitable for continuous or discrete numerical data
