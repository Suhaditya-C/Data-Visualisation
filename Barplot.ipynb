# Bar Plot


Defination: A bar plot (or bar chart) is a type of data visualization that uses rectangular bars to represent the frequency or value of categorical data. Each bar's length (or height) is proportional to the value it represents, making it easy to compare different categories



When to use: We use Bar plot to compare categorical data, to highlight differences

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
'''
matplotlib.pyplot.bar() takes some arguments:
x - The x coordinates of the bars
height - The height of the bars
width - The width of the bars
align - Alignment of the bars to the x coordinates
color - The color of the bars
edgecolor - the edgecolor of the bars
'''

#randint(a,b,n) gives n random integers between a and b
x=np.random.randint(10,90,30)
print(x)

# Create bins (intervals)
bins = np.arange(10, 91, 10)  # Bins from 10 to 90 with a step of 10

# Calculate the frequency of values in each bin
counts, bin_edges = np.histogram(x, bins=bins)

# Create the bar plot
plt.bar(bin_edges[:-1], counts, width=9, edgecolor='black')
plt.title("Bar Plot")
plt.xlabel("Values")
plt.ylabel("Frequency")
```

# Using Real Life Data

```python
#Importing data from CSV file using Pandas
data=pd.read_csv("MBA.csv")

#Storing Data of Interest
prof=data['work_industry']
#print(prof)

#To get the counts of repetitive professions
prof_count=prof.value_counts()
print(prof_count)

#Barplot and labels
sns.barplot(x=prof_count.index,y=prof_count.values,palette='rainbow');

plt.title("Bar Plot of Distribution of MBA Students by Profession")
plt.xlabel("Professions")
plt.ylabel("Count")
plt.xticks(rotation=45,ha='right')

plt.show()
```

From the barplot we can infer that:
* Consulting is the most common profession among MBA students
* Professions like CPG,Real Estate,Retail and Energy are the least represented with lower student counts

# Pros:
1. Bar plots are excellent for comparing different categories
2. Bar plots can be easily customized in terms of colors, orientations
3. Bar plots are particularly useful for displaying categorical data, allowing for easy visualization of frequency counts

# Cons:
1. Bar plots only convey information about the magnitude of categories. They do not show distribution, variability
2. Bar plot is less effective for small sample sizes
3. Bar plots assume that categories are independent of one another. If there is a relationship or correlation among categories, it may not be appropriately represented in a bar plot
