# Line Chart

Defination: A line chart (or line graph) is a type of data visualization that displays information as a series of data points connected by straight line segments. It is typically used to represent changes in values over a continuous variable, most often time

When to use: 

Line charts are ideal for visualizing data points across time periods, such as daily temperatures, monthly sales,
and when we need to observe how data behaves over intervals or continuous ranges, such as in stock market data


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

'''
For Line Chart plt.plot is enough cause plt.plot(x,y) takes 2 inputs where x and y being the x and y coordinates,
now plt.plot() just plots the corresponding points according to the data and , connects those points by a line, thats how it gives the line chart
'''

'''
Line chart is mainly used for time series data
so for example, lets x be the nth months and y be the average temp. in Bangalore
'''

#Data Generation
month=np.arange(1,13,1) 
temp=np.random.randint(15,28,12)

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

#Plot and Title
plt.plot(month,temp,marker='o',color='Orange')
plt.title('Line chart of Avg Temp vs Month in Bangalore')
plt.xlabel('Month')
plt.ylabel(r'Temperature ($\degree C$)')
plt.xticks(month,month_names)
plt.show()
```

# Using Real Life Data

```python
# Importing data from csv file using pandas
data = pd.read_csv("all_stocks_5yr.csv")
print(data.head(3))

# Convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Set the 'date' column as the DataFrame index
data=data.set_index('date')
print(data.head(3))

# Resample the data to monthly frequency, taking the mean of the 'close' prices for each month
# resample('M') resamples the data on monthly basis, likewise we can resample yearly basis as well
monthly_data = data['close'].resample('M').mean()

# Plot and Title
plt.plot(monthly_data.index, monthly_data.values, label='Monthly Close Price', color='red', linewidth=2,marker='.')
plt.title('Monthly Stock Price of American Airlines Over 5 Years')
plt.xlabel('Time Period')
plt.ylabel('Stock Price(USD)')
plt.grid(True)
plt.legend()
plt.show()
```

From the above line chart we can infer:
* The stock price shows a general upward trend
* The are noticable dips in stock during the time period of 2015-16
* The time period 2016-18 shows a strong growth


# Pros: 
1. Line charts are ideal for representing data with a continuous relationship, such as time series, temperatures
2. Line charts are excellent for displaying data trends, patterns, and changes over time or other continuous variables

# Cons: 
1. Line charts are not suitable for categorical or discrete data
2. If the x-axis data points (e.g., time intervals) are uneven or not uniformly spaced, the line chart might give a misleading representation of the trends
3. The visual impression of trends can be distorted if the axes are not properly scaled
