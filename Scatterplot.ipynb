# Scatter Plot

Defination : A scatter plot is a type of data visualization that displays the relationship between two continuous variables by plotting individual data points on a Cartesian plane.

When to use : Scatter plots are ideal for identifying the relationships or correlations between 2 variables and they also help to spot trends and cluster

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
'''
plt.scatter() takes some arguments:
x,y: data points or array of same size for plotting the point
c: For Color of a point/points
marker: How we want to represent the point
linewidths: Thickness/width of the point/marker
etc.
'''
# Generate synthetic data: 3000 points from a normal distribution
# Mean_1= 2,Sd_1= 3
# Mean_2= 4,Sd_2= 2
x = np.random.normal(2,3,3000)  # X values
y = np.random.normal(4,2,3000)  # Y values

# Scatter plot and Title
plt.scatter(x, y, marker='.',c='brown', alpha=0.5 ,linewidths=0.01)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xticks(np.arange(-10,17,4))
plt.title("Scatter Plot of Normally Distributed Data")
plt.axvline(2,color='black')
plt.axhline(4,color='black')

plt.show()
```

# Using Real Life Data

```python
# Importing data from CSV file using pandas
data=pd.read_csv("Spotify Most Streamed Songs.csv")
#print(data.head(3))

# Storing data of interest
dancibility=data['danceability_%'].to_numpy()
energy=data['energy_%'].to_numpy()


#Some datapoints
print(f'Some Values from Dancibility: {dancibility[10:20]}')
print(f'Some Values from Energy: {energy[10:20]}')

#Scatterplot using seaborn and title
sns.scatterplot(x=dancibility,y=energy,marker='o',c=np.random.rand(953),alpha=0.6)

plt.xlabel('Dancibility (in %)')
plt.ylabel('Energy (in %)')
plt.title('Scatter Plot of Dancibility and Energy')
#plt.plot(np.mean(dancibility),np.mean(energy),color='red',marker='x',label='Mean')
#plt.legend()

plt.show()
```

Scatter plot mainly tells about the spread and clustering of data.
So from the above plot we can infer that:  
* The majority of the points are clustered between 40 to 90 percent on energy scale and 40 to 80 percent in dancibility scale
* There are few outliers such as songs with high dancibility but low energy
* The data points are spread across the plot what appears to be uniform, suggesting that there is no clear linear correlation between them

# Pros:
1. Scatter plots are great for visually identifying the correlation between two continuous variables
2. Scatter plots make it easy to detect outliers
3. They can reveal clusters of data points or general trends

# Cons: 
1. Scatter plots are limited to visualizing the relationship between two variables at a time
2. Scatter plots don't directly provide summary statistics (like means, medians, or variability)
3. Scatter plots are designed for continuous data and are not suitable for visualizing categorical variables without modification
